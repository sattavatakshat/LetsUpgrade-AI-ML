```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop(["EmployeeID","EmployeeCount","Over18","StandardHours","StockOptionLevel"],axis = 1 , inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
data["BusinessTravel"] = l.fit_transform(data["BusinessTravel"])
data["Department"] = l.fit_transform(data["Department"])
data["EducationField"] = l.fit_transform(data["EducationField"])
data["Gender"] = l.fit_transform(data["Gender"])
data["JobRole"] = l.fit_transform(data["JobRole"])
data["MaritalStatus"] = l.fit_transform(data["MaritalStatus"])
data["Attrition"] = l.fit_transform(data["Attrition"])
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.isnull().sum()
```

```python
y = data["Attrition"]
x = data.drop("Attrition",axis = 1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .20)
```

```python
import statsmodels.api as sm
```

```python
x1 = sm.add_constant(x)
type(x1)
```

```python
log = sm.Logit(y, x1)
```

```python
model = log.fit()
```

```python
model.summary()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```
